```python
# 2️ Import Libraries
import pyspark
import pandas as pd

from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
# 3️ Create Spark Session
spark = SparkSession.builder \
        .appName("Customer_Churn_Prediction") \
        .getOrCreate()

print("Spark session created successfully!")
```

```python
# 4️ Upload Dataset (Colab)
from google.colab import files
uploaded = files.upload()
```

```python
# 5️ Load Dataset into PySpark DataFrame

df = spark.read.csv(
    "Telco-Customer-Churn.csv",
    header=True,
    inferSchema=True
)

print("Dataset Loaded Successfully!")
```

```python
# 6️ Show First 5 Rows
df.show(5)
```

```python
# 7️ Print Schema (Data Types)
df.printSchema()
```

```python
# 8️⃣ Check Number of Rows and Columns

print("Number of Rows :", df.count())
print("Number of Columns :", len(df.columns))
```

```python
# 9️⃣ View Column Names
df.columns
```

```python
# 🔟 View Data Types
df.dtypes
```

```python
# 11️⃣ Select Specific Columns
df.select("customerID", "tenure", "MonthlyCharges", "TotalCharges").show(5)
```

```python
# 12️⃣ Statistical Summary

df.describe().show()
```

```python
# 13️⃣ Check Null Values

from pyspark.sql.functions import isnan, when, count

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
# 14️⃣ Count Customers by Churn

df.groupBy("Churn").count().show()
```

```python
# 15️⃣ Count Customers by Contract Type

df.groupBy("Contract").count().show()
```

```python
# 16️⃣ Average Monthly Charges by Contract

from pyspark.sql.functions import avg

df.groupBy("Contract").agg(
    avg("MonthlyCharges").alias("Avg_Monthly_Charges")
).show()
```

```python
# 17️⃣ Add New Column (Example Feature)

df = df.withColumn(
    "MonthlyChargesPlus10",
    col("MonthlyCharges") + 10
)

df.select("MonthlyCharges", "MonthlyChargesPlus10").show(5)
```

```python
# 18️⃣ Drop Column Example

df = df.drop("MonthlyChargesPlus10")
```

```python
# 20️⃣ Sort Customers by Monthly Charges

df.orderBy(col("MonthlyCharges").desc()).show(5)
```
